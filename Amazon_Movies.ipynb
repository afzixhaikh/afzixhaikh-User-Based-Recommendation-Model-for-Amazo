```python
#now we preprocess the data for our recommendation model
data_melt=data.melt(id_vars= data.columns[0],value_vars=data.columns[1:],var_name='Movie',value_name='rating')
```

```python
data_melt
```

```python
#we fill in the nan values with 0
data_melt = data_melt.fillna(0)
```

```python
data_melt
```

```python
import surprise
from surprise import Reader
from surprise import Dataset
from surprise.model_selection import train_test_split
from surprise import SVD
```

```python
reader = Reader(rating_scale=(0,5))


data = Dataset.load_from_df(data_melt, reader=reader)
```

```python
Xtrain, Xtest = train_test_split(data, test_size=0.2)
```

```python
#We use Singular value decomposition as our model
svd = SVD()
```

```python
svd.fit(Xtrain)
```

```python
pred = svd.test(Xtest)
```

```python
from surprise.model_selection import cross_validate
```

```python
#We get a mean RMSE of 0.27
cross_validate(svd,data,measures=['RMSE'],cv=5,verbose=2)
```
